```python
cat = cat.drop(columns = ["Churn"])
```

```python
enc_cat = pd.get_dummies(cat, dtype = 'int64')
```

```python
enc_cat
```

```python
newtcc = num.join(enc_cat, lsuffix="_numeric" , rsuffix="_encoded_categorical dataframe series")
newtcc
```

```python
y
```

```python
x = newtcc
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,random_state = 42, test_size = 0.2)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
#scaler.fit(x_train)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 1000)
model
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
#Accuracy
Accuracy = accuracy_score(y_test, y_pred)
#confusion
cm = confusion_matrix(y_test, y_pred)
#precision
precision = precision_score(y_test, y_pred)
#f1
f1 = f1_score(y_test, y_pred)
#r2
r2 = r2_score(y_test, y_pred)
#recall
recall = recall_score(y_test, y_pred)

#print\
print('Accuracy :\n', Accuracy,'\n Precision Score: \n', precision, '\n f1 score: \n', f1 , '\n r2 score: \n',r2 , '\n recall score: \n',recall , '\n confusion matrix: \n', cm)
```

```python
#ROC
roc = roc_auc_score(y_test, y_pred)
roc
```

Here are the key areas that matter in a telecom report:
1. Financial Performance Metrics
Average Revenue Per User (ARPU): This is a primary indicator of operational performance, showing the average revenue generated per customer. High or growing ARPU, often driven by bundled services, indicates strong profitability.
Subscriber Growth (Net Additions): Measures how many new customers the company has added, which is essential for assessing future revenue potential.
Revenue Leakage & Billing Accuracy: Reports should detail revenue assurance efforts, particularly in identifying billing errors or fraudulent activity that causes revenue loss.
Operating Margin & Costs: Analysis of cost efficiency, especially managing operating expenses associated with maintaining vast infrastructure.
